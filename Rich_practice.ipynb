```python
import pandas as pd
import scipy.stats as scts
import seaborn as sns
```

```python
dw=pd.read_csv('Top_1000_wealthiest_people.csv')
dw
```

```python
group=dw['Net Worth (in billions)']
hypothesized_mean=50
tstats,p_value=scts.ttest_1samp(group,hypothesized_mean)
alpha=0.05
print(f'Tstatistic: {tstats}')
print(f'P_value: {p_value}')
alpha=0.05
if p_value<alpha:print("reject null hypothesis")
else:print("fail to reject null hypothesis")
```
