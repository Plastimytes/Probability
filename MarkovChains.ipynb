## MARKOV CHAINS
These are mathematical models that describe a sequence of events where the probability of a future event of a given system depends only on the current state of the system.<br>
This is an example of how it can be applied in jypter notebook

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Transition probability matrix
P=np.array([[0.7,0.3],[0.2,0.8]])
```

```python
#Initial state distribution
pI=np.array([0.6,0.4])
```

```python
#Function for Markov chain
def markov_chain(P,pI, num_steps=10):
    state_history=[np.random.choice([0,1],p=pI)]
    for i in range(num_steps):
        current_state=state_history[-1]
        next_state=np.random.choice([0,1],p=P[current_state])
        state_history.append(next_state)
    return state_history   

#Simulate chain for 100 steps
num_steps=100
state_history=markov_chain(P,pI,num_steps) 
```

```python
#Visualize state history
plt.plot(state_history)
plt.xlabel("Time step")
plt.ylabel("State")
plt.title("Markov Chain Simulation")
plt.show()
```
