```python
# Define events A and B
event_A={1,2,3}
event_B={2,3,4}
Total=len(event_A)+len(event_B)
Total
# Probability of event A 
Prob_A=len(event_A)/Total
Prob_B=len(event_B)/Total
print("The probability of event A is", Prob_A,"and the probability of event B is", Prob_B)

Joint=len(event_A & event_B)
Conditional_Prob=Joint/len(event_B)
print("The probability of A given B is",Conditional_Prob)
```
